## Замена нулей на медианные значения

```python
median_income = df['total_income'].median()
df['total_income'] = df['total_income'].replace(0, median_income)
df
```

```python
# (Возраст - 18) * 365
for i in range(len(df)):
    if df.loc[i, 'days_employed'] < 0:
        df.loc[i, 'days_employed'] = abs(df.loc[i, 'days_employed'])
    if df.loc[i, 'days_employed'] > 247 * 50:
        df.loc[i, 'days_employed'] = df.loc[i, 'days_employed'] / 24
```

```python
for i in range(len(df)):
    if df.loc[i, 'days_employed'] == np.nan and df.loc[i, 'days_employed'] > 0:
        df.loc[i, 'days_employed'] = (df.loc[i, 'days_employed'] - 18) * 247
        df['days_emloyed'] = df['days_employed'].fillna(df['days_employed'].median())
```

```python
for i in range(len(df)):
    if df.loc[i, 'dob_years'] == 0:
        df.loc[i, 'dob_years'] = df.loc[i, 'days_employed'] / 247
    if df.loc[i, 'dob_years'] > 50:
        df.loc[i, 'dob_years'] = df.loc[i, 'days_employed'] / 247
```
